Task 2: Build Advanced Preprocessing Function

```python
import re # For regex operations

def preprocess_text(text): # Preprocesses text for NLP tasks
    if not isinstance(text, str) or text.strip() == "": # Check if input is valid
        return [], "" # Return empty tokens and string for invalid input

    # 1. Lowercase
    text = text.lower() # Convert text to lowercase for uniformity

    # 2. Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text) # Remove URLs
    text = re.sub(r"\S+@\S+", "", text) # Remove email addresses

    # 3. Remove numbers
    text = re.sub(r"\d+", "", text) # Remove digits from text

    # 4. Remove emojis & special chars
    text = re.sub(r"[^\w\s]", "", text) # Remove punctuation and special characters

    # 5. Handle repeated characters (soooo → soo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text) # Reduce repeated characters to a single instance

    # 6. Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip() # Replace multiple spaces with a single space and trim

    # 7. Tokenization
    tokens = text.split() # Split text into tokens based on whitespace

    # 8. Remove short tokens (≤2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]] # Filter out short tokens while keeping negations

    cleaned_sentence = " ".join(tokens) # Join tokens back into a cleaned sentence

    return tokens, cleaned_sentence # Return list of tokens and cleaned sentence
```

Task 3: Stress Testing

```python
test_sentences = [ # Diverse test cases to evaluate preprocessing
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = [] # Store results for each test sentence

for sentence in test_sentences: # Process each test sentence through the preprocessing function
    tokens, clean = preprocess_text(sentence) # Get tokens and cleaned sentence from preprocessing

    print("\nOriginal:", sentence) # Print original sentence
    print("Tokens:", tokens) # Print list of tokens after preprocessing
    print("Clean Sentence:", clean) # Print cleaned sentence after preprocessing

    results.append((tokens, clean)) # Append results to the list for further analysis if needed
```

Task 4: Token Analytics

```python
import numpy as np

for i, (tokens, _) in enumerate(results): # Analyze token statistics for each sentence
    if len(tokens) == 0: # Skip sentences that resulted in no tokens after preprocessing
        continue # Avoid division by zero and meaningless statistics

    total_tokens = len(tokens) # Total number of tokens in the sentence
    unique_tokens = len(set(tokens)) # Number of unique tokens in the sentence
    avg_len = np.mean([len(t) for t in tokens]) # Average token length in the sentence

    print(f"\nSentence {i+1}:") # Print sentence index for clarity
    print("Total Tokens:", total_tokens) # Print total number of tokens
    print("Unique Tokens:", unique_tokens) # Print number of unique tokens
    print("Average Token Length:", round(avg_len, 2)) # Print average token length
```

Task 5: Frequency Analysis

```python
from collections import Counter # For counting token frequencies

all_tokens = [] # Collect all tokens from all sentences for frequency analysis

for tokens, _ in results: # Extend the all_tokens list with tokens from each sentence
    all_tokens.extend(tokens) # Add tokens from current sentence to the overall list of tokens

counter = Counter(all_tokens) # Create a Counter object to count the frequency of each token

print("\nTop 10 Frequent Words:") # Print the 10 most common tokens and their counts
print(counter.most_common(10)) # Print the 10 most common tokens along with their frequency counts

print("\nTop 5 Least Frequent Words:") # Print the 5 least common tokens and their counts
print(counter.most_common()[-5:]) # Print the 5 least common tokens along with their frequency counts (last 5 in the list)
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list): # Process a list of texts through the full preprocessing pipeline and return tokens and cleaned sentences
    all_tokens = [] # Collect all tokens from all sentences
    clean_sentences = [] # Collect cleaned sentences for output

    for text in text_list: # Process each text in the input list through the preprocessing function
        tokens, clean = preprocess_text(text) # Get tokens and cleaned sentence from preprocessing
        all_tokens.extend(tokens) # Add tokens from current sentence to the overall list of tokens
        clean_sentences.append(clean) # Add cleaned sentence to the list of clean sentences

    return {
        "tokens": all_tokens, # Return the combined list of all tokens from all sentences
        "clean_sentences": clean_sentences # Return the list of cleaned sentences corresponding to the input texts
    }
```
